Mammogram Stats

```python
dfAbove40 = df[(df['Age'] != 1) & (df['Age'] != 2)]

print("Had a mammogram:")
print("White")
print(dfAbove40[dfAbove40['Q52']=='5'].Q11.value_counts())
print("Hispanic")
print(dfAbove40[dfAbove40['Q52']=='3'].Q11.value_counts())
print("Black")
print(dfAbove40[dfAbove40['Q52']=='2'].Q11.value_counts())

#dfAbove40.Q13.value_counts()
print("Compliance:")
print("White")
print(dfAbove40[dfAbove40['Q52']=='5'].Q13.value_counts())
print("Hispanic")
print(dfAbove40[dfAbove40['Q52']=='3'].Q13.value_counts())
print("Black")
print(dfAbove40[dfAbove40['Q52']=='2'].Q13.value_counts())
```

```python
#print("Overview of Incomes:")
#dfAbove40.HouseholdIncome.value_counts()

# More than 40k, and less than 40k - two groups

print("Mammogram stats:")
print("Less than 40k income - mammogram stats:")
print(dfAbove40[dfAbove40['HouseholdIncome'] <=2].Q11.value_counts())
print("More than 40k income - mammogram stats:")
print(dfAbove40[dfAbove40['HouseholdIncome'] > 2].Q11.value_counts())

# Compliance Stats

print("Compliance stats:")
print("Less than 40k income - comp stats:")
print(dfAbove40[dfAbove40['HouseholdIncome'] <=2].Q13.value_counts())
print("More than 40k income - comp stats:")
print(dfAbove40[dfAbove40['HouseholdIncome'] > 2].Q13.value_counts())
```

```python
#print("Overview of Education Levels")
#dfAbove40.Q3.value_counts()

# High School or below, More than high school - two groups

print("Mammogram Stats:")
print("Less than HS - mammogram stats:")
print(dfAbove40[dfAbove40['Q3'] <=2].Q11.value_counts())
print("More than HS - mammogram stats:")
print(dfAbove40[dfAbove40['Q3'] > 2].Q11.value_counts())

# Compliance Stats
print("Comp Stats:")
print("Less than HS - mammogram stats:")
print(dfAbove40[dfAbove40['Q3'] <=2].Q13.value_counts())
print("More than HS - mammogram stats:")
print(dfAbove40[dfAbove40['Q3'] > 2].Q13.value_counts())
```
